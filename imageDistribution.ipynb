```python
from utils import load_tang, load_cifar
import numpy as np
import matplotlib.pyplot as plt
import torchvision.transforms as transforms
```

```python
trainImgs, valImgs = load_tang()
```

```python
trainCIFAR, valCIFAR = load_cifar()
```

```python
transformTang=transforms.Compose([
                                 transforms.ToTensor(),
                                 transforms.Normalize(0.5, 0.5),

                             ])
```

```python
transformedCifar = np.array([trainCIFAR.transform(i).numpy() for i in trainCIFAR.data])
```

```python
transformedTang = np.array([transformTang(i).numpy() for i in trainImgs.data.numpy()])
```

```python
histogram, bin_edges = np.histogram(transformedCifar, bins=256)
histogramT, bin_edgesT = np.histogram(transformedTang, bins=256)
```

```python
plt.figure()
plt.title("Grayscale Histogram")
plt.xlabel("grayscale value")
plt.ylabel("pixel count")
#plt.xlim([0.0, 1.0])  # <- named arguments do not work here

plt.plot(bin_edges[0:-1], histogram)
plt.figure()
plt.title("Grayscale Histogram")
plt.xlabel("grayscale value")
plt.ylabel("pixel count")
#plt.xlim([0.0, 1.0])  # <- named arguments do not work here

plt.plot(bin_edgesT[0:-1], histogramT)
```
